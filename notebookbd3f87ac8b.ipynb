# 23. Average Rating per Genre

```python
all_genres = []
for index, row in data.iterrows():
    genres = row['Genre'].split(',')
    for genre in genres:
        all_genres.append(genre.strip())

unique_genres = pd.Series(all_genres).unique()
print(f"Number of unique genres: {len(unique_genres)}")
print(f"Unique genres: {unique_genres}")
```

```python
genre_ratings = {}
for genre in unique_genres:
    movies_in_genre = data[data['Genre'].str.contains(genre, case=False, na=False)]
    if not movies_in_genre.empty:
        genre_ratings[genre] = movies_in_genre['Rating'].mean()
avg_rating_per_genre = pd.DataFrame(genre_ratings.items(), columns=['Genre', 'Average Rating'])
avg_rating_per_genre = avg_rating_per_genre.sort_values(by='Average Rating', ascending=False)

display(avg_rating_per_genre.head())
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Average Rating', y='Genre', data=avg_rating_per_genre.head(10), palette='viridis')
plt.title('Top 10 Genres by Average Movie Rating')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 24. Correlation Between Movie Runtime and Rating

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runtime (Minutes)', y='Rating', data=data, alpha=0.6)
plt.title('Movie Runtime vs. Rating')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

correlation = data['Runtime (Minutes)'].corr(data['Rating'])
print(f"Correlation between Runtime (Minutes) and Rating: {correlation:.2f}")
```

# 28. Top Actors by Number of Movies

```python
all_actors = []
for index, row in data.iterrows():
    actors = row['Actors'].split(',')
    for actor in actors:
        all_actors.append(actor.strip())

actor_counts = pd.Series(all_actors).value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=actor_counts.head(10).values, y=actor_counts.head(10).index, palette='magma')
plt.title('Top 10 Actors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Actor')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
